# Eigendecomposition

$M x = \lambda x \implies (\lambda, x)$ is an eigenpair of the square matrix $M$


A square matrix $M$ can be factorized as $M = CDC^{-1}$ where $D$ is a diagonal matrix 

**Spectral Theorem**: For a real, symmetric matrix $S$,
$S = Q \Lambda Q^T$ 

$S = \lambda_1 q_1 q_1^T + \cdots + \lambda_n q_n q_n^T$

```python
M = np.array([[2, 1],[1, 2]])
```

```python
evals, evecs = np.linalg.eigh(M)
```

```python
eidx = np.argsort(evals)[::-1]
evecs = evecs[:,eidx]
evals = evals[eidx]
```

```python
D = np.diag(evals)
```

```python
D
```

```python
C = evecs
C
```

```python
plot_eigen(M)
```

```python
C_inv = np.linalg.inv(C)
```

```python
plot_linear_transformation(C_inv,x,unit_circle=True)
```

```python
plot_linear_transformation(D,x,unit_circle=True)
```

```python
plot_linear_transformation(C,x,unit_circle=True)
```

```python
plot_linear_transformations(C_inv,D,C, unit_circle=True)
```

```python
assert np.allclose(M @ C[:,0], D[0,0] * C[:,0])
```

```python
assert np.allclose(M @ C[:,1], D[1,1] * C[:,1])
```
